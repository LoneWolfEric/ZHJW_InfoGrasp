```python
img = Image.open('captcha2.jpg')
img = np.asarray(img)
img = np.array(img)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if ( img[i,j,0] > 80 ) and ( img[i,j,1] < 140 ) and ( img[i,j,2] < 140 ):
            img[i,j] = np.array([0,0,0])
        else:
            img[i,j] = np.array([255,255,255])

plt.imshow(img)
```
